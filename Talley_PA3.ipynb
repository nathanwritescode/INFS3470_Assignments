```python
# (1) import the pandas module
import pandas as pd
```

```python
# use pandas to read in the file 'Pokemon.csv' into a dataframe
# (2) display the first 10 rows of the dataframe
df = pd.read_csv('Pokemon.csv')
df.head(10)
```

```python
# (3) filter the dataframe on only the 'number', 'name', 'type1', and 'type2' columns
# display the first 10 rows
df[ ['number', 'name', 'type1', 'type2']].head(10)
```

```python
# (4) filter the dataframe to only display rows where the 'speed' value is 150
df[ (df['speed'] > 150)]
```

```python
# (5) filter the dataframe to display only the 'name' column where 'atk' is 100
df[ ['name'] ][(df['atk'] == 100)]
```

```python
# (6) Answer the question: How many pokemon have a 'def' value greater than or equal to 100?
len(df[(df['def'] >= 100)])
```

```python
# (7a) What is the average 'atk' value of 'type1' psychic pokemon?
# (7b) What is the average 'atk' value of 'type1' fighting pokemon?
# Hint: use .mean()

print("The average atk value of psychic pokemon is %.2f" %df[(df['type1'] == 'psychic')]['atk'].mean())
print("The average atk value of fighting pokemon is %.2f" %df[(df['type1'] == 'fighting')]['atk'].mean())
```

```python
# (8a) What is the most common 'type1' for pokemon with a 'name' starting with 'a'?
# (8b) What is the most common 'type1' for pokemon with a 'name' starting with 'p'?
dfa = df[df.name.str[0] == 'a']
print("The most common type1 for pokemon starting with 'a' is %s" %dfa.type1.mode()[0])
dfp = df[df.name.str[0] == 'p']
print("The most common type1 for pokemon starting with 'p' is %s" %dfp.type1.mode()[0])
```

```python
# (9) List the 'name's of all pokemon with a 'def' value that is at least four times greater 
# than 'hp' value
df[ ((df['def']) > df['hp'] * 4)]['name']
```

```python
# (10) How many pokmeon are either 'type1' or 'type2' ghost?
len(df[(df['type1'] == 'ghost') | (df['type2'] == 'ghost')]['name'])
```
